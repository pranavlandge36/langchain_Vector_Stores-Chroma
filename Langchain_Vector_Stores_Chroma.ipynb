```python
from langchain_core.documents import Document
from langchain_openai.embeddings import OpenAIEmbeddings
from langchain_chroma import Chroma
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
embeddings = OpenAIEmbeddings(
    model="openai/text-embedding-ada-002",  # or any model from OpenRouter
    api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1",
    # streaming=True
)
```

```python

documents = [
    Document(
        page_content="A genius billionaire who builds advanced armored suits to fight threats.",
        metadata={"name": "Iron Man"}
    ),
    Document(
        page_content="A super soldier enhanced with a serum, known for his shield and leadership.",
        metadata={"name": "Captain America"}
    ),
    Document(
        page_content="A Norse god of thunder who wields a magical hammer and controls lightning.",
        metadata={"name": "Thor"}
    ),
    Document(
        page_content="A brilliant scientist who transforms into a powerful green giant when angry.",
        metadata={"name": "Hulk"}
    ),
    Document(
        page_content="A skilled spy and assassin with exceptional combat abilities and intelligence.",
        metadata={"name": "Black Widow"}
    )
]
```

```python
documents
```

```python
vector_store=Chroma(
    embedding_function=embeddings,
    persist_directory="db",
    collection_name='mcu_characters'

)
vector_store.add_documents(documents)
vector_store
```

```python
print(vector_store.get(include=['embeddings','documents','metadatas']))
```

```python
vector_store.similarity_search(
    query='who uses hammer?',
    k=1
)
```

```python
# search with similarity search
vector_store.similarity_search_with_score(
    query='who is called green giant',
    k=1
)
```

```python
vector_store.similarity_search_with_score(
    query='',
    filter={'name':'Iron Man'},
    k=1)
```

```python
update_doc=Document(
    page_content=' A character from MCU, also known as god of thunder , weilds hammer and controls lightning.',
    metadata={'name':'Thor'}
)

vector_store.update_document(document_id='b4eecb25-a2b9-4e48-9f3e-9f4ccb15d2ce',document=update_doc)
```
